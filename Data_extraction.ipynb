1.  Describe different types of data sources used in ETL with suitable examples.

    - In ETL, data sources are the systems or locations from which data is extracted before being transformed and loaded into a data warehouse or database. The main types of data sources are:
    - Relational Databases (RDBMS), Flat Files, Enterprise Applications, Web Services / APIs, Cloud Storage.
    - Relational databases – Customer database in MySQL.
      Flat files – CSV or Excel sales reports.
      Enterprise applications – SAP ERP, Salesforce CRM.
      Web APIs – Online weather or payment APIs.
      Cloud storage – Amazon S3 or Azure Blob Storage.

2. What is data extraction? Explain its role in the ETL pipeline.

    - Data extraction is the first step in the ETL (Extract, Transform, Load) process. It involves collecting data from one or more source systems, such as databases, files, cloud storage, APIs, or enterprise applications, and preparing it for transformation and loading into a target system like a data warehouse.
    - Role of Data Extraction in the ETL Pipeline:-
      Collects data from multiple sources, Provides input for transformation, Supports data integration, Ensures data quality, Improves business reporting.

3.  Explain the difference between CSV and Excel in terms of extraction and ETL usage.

    - CSV files are simple, lightweight, and ideal for fast ETL processing of structured data.
     Excel files support advanced features like formulas, formatting, and multiple worksheets, making them suitable for reporting but slightly more complex to extract.
    - CSV: A retail store exports daily transaction data as sales.csv, which is directly loaded into a data warehouse.
      Excel: An HR department maintains employee details in Employees.xlsx with multiple sheets (Employee Info, Payroll, Attendance), which are extracted separately during ETL.

4.  Explain the steps involved in extracting data from a relational database.

    - Data extraction from a relational database is the process of retrieving data from database tables for use in an ETL (Extract, Transform, Load) pipeline. The main steps are:-
    - Connect to the Database, Identify the Required Data, Extract the Data, Validate the Extracted Data, Transfer to the Transformation Stage.
    - Extracting data from a relational database involves connecting to the database, selecting the required data, executing SQL queries, validating the extracted data, optionally storing it in a staging area, and sending it to the transformation stage. This process ensures that accurate and reliable data is available for ETL and business analytics.

5.  Explain three common challenges faced during data extraction.

    - Data Quality Issues:- Source data may contain missing values, duplicate records, incorrect formats, or inconsistent data.
     Poor-quality data can lead to inaccurate analysis and reporting.
    - Data Integration from Multiple Sources:- Organizations often extract data from different systems such as databases, Excel files, APIs, and cloud storage.These sources may have different formats, structures, and naming conventions, making integration difficult.
    - Performance and Large Data Volumes:- Extracting very large datasets   can be time-consuming and may slow down the source database.Network delays and limited system resources can also affect extraction performance.


6.  What are APIs? Explain how APIs help in real-time data extraction.

    - API (Application Programming Interface) is a set of rules and protocols that allows different software applications to communicate and exchange data with each other. In ETL, APIs are used to extract data directly from online applications, cloud services, or web platforms without accessing the underlying database.
    * How APIs Help in Real-Time Data Extraction:-
    - Provides Live Data, Enables Automatic Data Retrieval, Integrates Multiple Systems, Reduces Manual Work, Supports Real-Time Analytics.

8.  What steps should an ETL developer take when extracting data from large CSV files (1GB+)?

   - When working with very large CSV files, an ETL developer should follow these best practices to ensure efficient and reliable extraction:
   - Validate the CSV File, Read the File in Chunks, Filter Only Required Data, Use Parallel Processing (If Supported), Perform Basic Data Validation.
   - Validate the file format.
     Read the file in batches of 50,000 rows.
     Extract only required columns (OrderID, CustomerID, SalesAmount).
     Validate data and remove duplicates.
     Store the extracted data in a staging table.
